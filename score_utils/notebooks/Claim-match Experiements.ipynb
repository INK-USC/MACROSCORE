## Get the PageNumber and Coordinates for each Textbox element in PDF:

```python
from pdfminer.layout import LAParams, LTTextBox, LTTextLine
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator

def getPageLayoutVals(cur_pdf_file_name):
    result = {}
    fp = open(cur_pdf_file_name, 'rb')
    
    # PDFMiner Configs
    rsrc_mgr = PDFResourceManager()
    la_params = LAParams()
    device = PDFPageAggregator(rsrc_mgr, laparams=la_params)
    interpreter = PDFPageInterpreter(rsrc_mgr, device)
    pages = PDFPage.get_pages(fp)
    
    cur_page_num = 0
    for page in pages:
        cur_page_num += 1
        
        # Insert the page number to result
        if result.get(cur_page_num) is None:
            result[cur_page_num] = []
        
        interpreter.process_page(page)
        layout = device.get_result()
        for l_obj in layout:
            if isinstance(l_obj, LTTextBox):
                for txt_obj in l_obj:
                    if isinstance(txt_obj, LTTextLine):
                        cur_coords = [round(x,3) for x in txt_obj.bbox]
                        cur_text = txt_obj.get_text()
#                         print(cur_text)
                        
                        result[cur_page_num].append({
                            "text": str(cur_text.strip()),
                            "coords": tuple(cur_coords)
                        })
    
    return result
```

## Map the PageNumber and Coordinates for each sentence/heading in PDF:

```python
def normalizeText(input_text, remove_spaces=True):
    # Lowercase and retain only alpha-numeric characters and underscore (_)
    result = " ".join(re.sub("\W", " ", input_text.lower()).split())
    
    # Replace known greek symbols to corresponding english alphabets: (Eg: "ﬁ" to "fi" etc)
    result = unidecode(result)
    result = result.lower()
    
    # Remove remaining unicode characters (greek symbols)
    result_cleaned = ""
    for char in result:
        if char == " " or char.isnumeric() or (ord(char) >= ord('a') and ord(char) <= ord('z')):
            result_cleaned += char

    # Remove spaces
    if remove_spaces:
        result_cleaned = result_cleaned.replace(" ", "")
    
    return result_cleaned

def mapPageLayoutwithParsedContent(cur_parsed_json_file_name, cur_page_layout_file_name):
    try:
        with open(cur_parsed_json_file_name, "r") as f:
            parsed_data = json.load(f)

        with open(cur_page_layout_file_name, "r") as f:
            cur_page_layout_info = json.load(f)
    except Exception as e:
        print(e)
        return None
        
    # Get Section Heading <-> Page Number mapping
    
    # Get unique section headings:
    sec_head = {}
    sec_head["abstract"] = set()
    for cur_line in parsed_data["pdf_parse"]["body_text"]:
        cur_sec = cur_line["section"]
        if len(cur_sec) > 0 and sec_head.get(cur_sec.lower()) is None:
            sec_head[cur_sec.lower()] = set()
            
    # Map the section headings with page_number vals
    for cur_sec_head in sec_head.keys():
        cur_sec_head_norm = normalizeText(cur_sec_head)
        
        is_cont = False
        for cur_page_num in cur_page_layout_info.keys():
            cur_sent_list = cur_page_layout_info[cur_page_num]
            for cur_sent in cur_sent_list:
                cur_text = cur_sent["text"]
                
                if is_cont:
                    cur_text_norm += normalizeText(cur_text)
                else:
                    cur_text_norm = normalizeText(cur_text)
                    
                if len(cur_text_norm) < 1:
                    continue

                if cur_text_norm == cur_sec_head_norm:
                    sec_head[cur_sec_head].add(cur_page_num)
                elif cur_text_norm in cur_sec_head_norm:
                    is_cont = True
                else:
                    is_cont = False
                
    return sec_head
```

```python
pdf_filename_mappings_path = "data/SCORE_csv.csv"
pdf_filename_mappings_df = pd.read_csv(pdf_filename_mappings_path)
```

```python
doi_pdf_file_map = {}
for idx, cur_row in pdf_filename_mappings_df.iterrows():
    cur_doi = cur_row["DOI_CR"]
    cur_pdf_file = cur_row["pdf_filename"]
    doi_pdf_file_map[cur_doi] = cur_pdf_file
```

```python
doi_heading_map = {}
```

```python
s2orc_path = "../s2orc-doc2json/outputs/{}.json"
pdfminer_path = "../s2orc-doc2json/pdfminer_outputs/{}.json"

for cur_doi, cur_pdf_file in tqdm(doi_pdf_file_map.items()):
    cur_pdf_pref = cur_pdf_file.split(".pdf")[0]
    cur_s2orc_path = s2orc_path.format(cur_pdf_pref)
    cur_pdfminer_path = pdfminer_path.format(cur_pdf_pref)
    cur_out = mapPageLayoutwithParsedContent(cur_s2orc_path, cur_pdfminer_path)
    if cur_out is not None:
        doi_heading_map[cur_doi] = cur_out
```

```python
len(doi_heading_map.keys())
```

```python
num_headings_list = []
num_total_headings_list = []
for k, v in doi_heading_map.items():
    cnt = 0
    for kk, vv in v.items():
        if len(vv) > 0:
            cnt += 1
    num_headings_list.append(cnt)
    num_total_headings_list.append(len(v.keys()))
```

```python
percent_headings_list = []
for x, y in zip(num_headings_list, num_total_headings_list):
    percent_headings_list.append(x/y)
```
